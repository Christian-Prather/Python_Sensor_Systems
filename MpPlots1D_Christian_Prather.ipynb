# Definition
**Matplotlib** is a visualization package for **numpy** arrays. Matplotlib can be used directly in Jupyter notebooks by producing embedded figures. The **pyplot** module provides a MATLAB-like interface.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10*np.pi, 1000)

ys = np.sin(x)
yc = np.cos(x)
```

# plt.plot()
Plot y versus x as lines and/or markers.

`plot([x], y, [fmt], data=None, **kwargs)`

* The coordinates of the points in the graph are given by **x** and **y**.
* **fmt** defines basic formatting like color, marker and linestyle.

```python
plt.plot(x,ys)
```

```python
# dashed line
plt.plot(x,ys,'--')
```

```python
# dotted line
plt.plot(x,ys,':')
```

```python
# line of points
plt.plot(x[::5],ys[::5],'.')
```

```python
# multiple lines; colors
plt.plot(x,ys,'r') # red
plt.plot(x,yc,'b') # blue
```

# plt.axis()
Get or set axis properties.

`plt.axis(*v, **kwargs)`

* turn off the axis lines and labels <br>
`axis('off')`

* changes limits so that equal increments of x and y have the same length <br>
`axis('equal')`

* changes axes limits such that all data are shown <br>
`axis('tight')`

* changes axes limits such that the have the same scaling <br>
`axis('square')`

* sets the axes limits to the values defined by the min/max pairs <br>
`axis( [xmin, xmax, ymin, ymax] )`

```python
# multiple lines; colors
plt.plot(x,ys,'r')
plt.plot(x,yc,'b')
plt.axis([2*np.pi, 6*np.pi, -0.5, +0.5]);
```

# plt.xlim(), plt.ylim()
Get or set the x limits of the current axes.
The min and max values are defined with a tuple.

`plt.xlim(*args, **kwargs)`

`plt.ylim(*args, **kwargs)`

```python
plt.plot(x,ys,'r')
plt.plot(x,yc,'b')

plt.xlim(5,15)
plt.ylim(0,1)
```

# plt.subplot()
Return a subplot axes at the given grid position.

`subplot(nrows, ncols, index, **kwargs)`

```python
# first panel
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, ys,'r')

# second panel
plt.subplot(2, 1, 2)
plt.plot(x, yc,'b');
```

```python
# first panel
plt.subplot(1, 2, 1) # (rows, columns, panel number)
plt.plot(x, ys,'r-')

# second panel
plt.subplot(1, 2, 2)
plt.plot(x, yc,'b--');
```

# line attributes

* color
* style
* width

```python
# first panel
plt.subplot(1, 2, 1) # (rows, columns, panel number)
plt.plot(x, ys, color='r',linestyle='-',linewidth=1)

# second panel
plt.subplot(1, 2, 2)
plt.plot(x, yc, color='b',linestyle='--',linewidth=3);
```

# labels and title
Set the label of a specific axis.

`plt.xlabel(s, *args, **kwargs)`

`plt.ylabel(s, *args, **kwargs)`

Set tht title of a plot.

`plt.title(s, *args, **kwargs)`

```python
plt.plot(x, ys,'k-')
plt.title('sin function')
plt.xlabel('a[radians]')
plt.ylabel('sin(a)')
```

# uncertainty bars
Plot y versus x as lines and/or markers with attached uncertainty bars.

`plt.errorbar(x, y, yerr=None, xerr=None, fmt='', ecolor=None, elinewidth=None, capsize=None, barsabove=False, lolims=False, uplims=False, xlolims=False, xuplims=False, errorevery=1, capthick=None, hold=None, data=None, **kwargs)`

```python
e = 0.1 + 0.25* np.abs(ys)
plt.errorbar(x[::15], ys[::15], yerr=e[::15], fmt='o');
```

# logarithmic plots
Make a plot with log scaling on one or more axes.

`plt.semilogx(*args, **kwargs)`

`plt.semilogy(*args, **kwargs)`

`plt.loglog(*args, **kwargs)`

```python
y = np.exp(0.2*x) * (1+np.sin(x))

plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.semilogy(x,y)

plt.subplot(2, 2, 3)
plt.semilogx(x,y)

plt.subplot(2, 2, 4)
plt.loglog(x,y)
```

# legend
Places a legend on the axes.

`plt.legend(*args, **kwargs)`

```python
plt.plot(x, ys,'r')
plt.plot(x, yc,'b');
plt.legend(['sin','cos'])
```

# grid lines
Turn the axes grids on or off.

`plt.grid(b=None, which='major', axis='both', **kwargs)`

```python
plt.plot(x, ys,'r')
plt.plot(x, yc,'b');
plt.legend(['sin','cos'])

plt.grid(True)
```

# ticks
Get or set the current tick locations and labels on a given axis.

`plt.xticks(*args, **kwargs)`

`plt.yticks(*args, **kwargs)`

```python
plt.plot(x, ys,'r')
plt.plot(x, yc,'b');

plt.grid(True)
plt.xticks([2,11,21])
plt.yticks([-0.5,-0.1,0.8])
```

# scatter plots
A scatter plot of y vs x with specified marker size and/or color.

`plt.scatter(x, y, **kwargs)`

```python
plt.plot(   ys[::10],yc[::10],'r')
plt.scatter(ys[::10],yc[::10])

plt.axis('equal')
plt.axis('off');
```
