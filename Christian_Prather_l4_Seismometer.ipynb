```python
# Include needed libraries such as i2c and spi drivers
import i2cUtils as ic
import numpy as np
import matplotlib.pyplot as plt
import time
import i2cUtils as ic
import RPi.GPIO as GPIO
import math as calc
import spiUtils as su



# Use GPIO numbering
GPIO.setmode(GPIO.BCM)

# Assign a variable for the tilt switch pin number
switchPin = 24

# Configures the GPIO pin to accept inputs
# Set the tilt switch to be a pull up configuration (logic high)
GPIO.setup(switchPin, GPIO.IN, pull_up_down=GPIO.PUD_UP)

#Debuging
#print(GPIO.input(switchPin))

# Accelerometer class for storing accelerometer values and defaults
class Accelerometer:
    # Default all accelration to zero so as to not break at read
    def __init__(self,xVal = 0,yVal = 0,zVal = 0,mag = 0): 
        self.X = xVal
        self.Y = yVal
        self.Z = zVal
        self.Mag = mag

    # Function for reading values used by all other functions
    def readAcceleration(self):
        self.X,self.Y,self.Z = ic.readACC()

    # Functions for x,y,z,mag calculations and readings
    def getAcceleration(self):
        self.readAcceleration()
        return self.X, self.Y, self.Z
        
    
    def getAccelerationX(self):
        #self.readAcceleration()
        return self.X


    def getAccelerationY(self):
        #self.readAcceleration()
        return self.Y

    def getAccelerationZ(self):
        #self.readAcceleration()
        return self.Z

    def getMagnitude(self):
        return calc.sqrt(pow(self.X,2)+pow(self.Y,2)+pow(self.Z,2))
        
#TODO: Find a better way of allocation memory for the numpy arrays - clp
nDATA = 200000

# Allocate arrays in memory for storing data
xACC = np.zeros(nDATA,dtype='float')
yACC = np.zeros(nDATA,dtype='float')
zACC = np.zeros(nDATA,dtype='float')
mACC = np.zeros(nDATA,dtype='float')
tACC = np.zeros(nDATA,dtype='float')
dPIEZO = np.zeros(nDATA,dtype='float')

# Voltage conversion for piezo data
toVolt = 3.3/1023

# Initiate an Accelerometer object 
acclerometer = Accelerometer()

# Until we break out check switch state and recorde data if triggered
while True:
    state = GPIO.input(switchPin) # TILT switch
    
    #print(state)
    if state: # If tilt switch active get data
        # Accelerometer Data
        print("Shaky Earth :(")
        
        # Time of start
        ti = time.time()
        # Incrementer
        i = 0
        
        # Recorde data until 15 seconds has passed
        while (time.time() - ti) <= 15:
           
            dPIEZO[i] = toVolt * su.readADC(channel=0) # PIEZO element
            tACC[i] = time.time() - ti
            acclerometer.getAcceleration()
            #print(acclerometer.getAcceleration())
            
            # Save accelerations to respective arrays
            xACC[i] = acclerometer.getAccelerationX()
            yACC[i] = acclerometer.getAccelerationY()
            zACC[i] = acclerometer.getAccelerationZ()
            mACC[i] = acclerometer.getMagnitude()
            i +=1
        # Break out of while true loop
        break
        
print("Recorded...")           

# Save data to File
np.savez('flatEarth.npz', tACC,xACC,yACC,zACC,mACC,dPIEZO) 
print("Saved to flatEarth.npz")

# Plot X,Y,Z,Piezo data vs time
plt.plot(tACC,xACC,'r')
plt.title("Time vs. X Acceleration")
plt.xlabel("Time")
plt.ylabel("X Acceleration")
plt.show()
plt.plot(tACC,yACC,'g')
plt.title("Time vs. Y Acceleration")
plt.xlabel("Time")
plt.ylabel("Y Acceleration")
plt.show()
plt.plot(tACC,zACC,'b')
plt.title("Time vs. Z Acceleration")
plt.xlabel("Time")
plt.ylabel("Z Acceleration")
plt.show()
plt.plot(tACC,dPIEZO,'k')
plt.title("Time vs. Piezzo Voltage")
plt.xlabel("Time")
plt.ylabel("Piezzo Voltage")
plt.show()# Write your code here!
```
